```python
import pandas as pd
import polars as pl 

# pc crashed when i tried to load file as json lmao so converted to parquet for optimization
#(
#    pl.scan_ndjson('season_10_20260427_070645.jsonl', infer_schema_length=None)
#    .sink_parquet('matches.parquet')
#)

all_matches = pd.read_parquet('matches.parquet')
all_matches
```

```python
print(all_matches.columns)
```

```python
# getting all ranked matches
ranked_matches = all_matches[all_matches['type'].isin([2])]
ranked_matches.info()
```

```python
# removing games that were created for decay
ranked_matches = ranked_matches[~ranked_matches['decayed']]

# dropping unneccessary columns for simplicity (features values are either always the same for ranked games or not useful in our investigation)
ranked_matches_cleaned = ranked_matches.drop(columns=['tag', 'vod', 'date', 'rank', 'type', 'season', 'decayed', 'beginner', 'category', 'gameMode', 'botSource', 'spectators', 'replayExist'])
ranked_matches_cleaned
```

```python
ranked_matches_cleaned.info()
```

```python
ranked_matches_cleaned.to_parquet('rankedmatches.parquet')
```
